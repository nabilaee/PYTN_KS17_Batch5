### Ploting in Pandas

```python
haiti = df_can.loc['Haiti', years]
haiti.head()
```

```python
haiti.plot()
```

```python
#memberi label sumbu x dan y
haiti.index = haiti.index.map(int) #mengubah nilai index ke bentuk int untuk plotting
haiti.plot(kind='line')
plt.title('Immigration from Haiti')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show() #untuk menampilkan update yang dibuat ke dalam gambar
```

```python
# memeberi anotasi lonjakan menggunakan metode plt.text()
haiti.plot(kind='line')

plt.title('Imigration from Haiti')
plt.ylabel('Number of Imigrants')
plt.xlabel('Years')

plt.text(2000, 6000, '2010 Earthquake')

plt.show()
```

```python
# menambahkan data ke plot garis untuk membuat perbandingan
#menampilkan data china dan india
China_India = df_can.loc[['China', 'India'], years]
China_India
```

```python
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df_top5 = df_can.head(5)
df_top5 = df_top5[years].transpose()
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='line', figsize=(14, 8))

plt.title('Imigration trend of top 5 countries')
plt.ylabel('Number of Imigrants')
plt.xlabel('Years')

plt.show()
```

### Area Plot

```python
# cumulaive plot
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df_top5 = df_can.head()
df_top5 = df_top5[years].transpose()

df_top5.head()
```

```python
#untuk menghasilkan stacked area plot setiap kolom harus semua poritif atau semua negatif
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind ='area',
             alpha = 0.25,
            stacked = False,
            figsize=(20, 10)
            )
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Imigrants')
plt.xlabel('Years')

plt.show()
```

```python
#scripting layer
df_top5.plot(kind='area', alpha = 0.35, figsize=(20, 10))
plt.title('Immigration trend of top 5 countries')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
```

```python
#Artist Layer
ax = df_top5.plot(kind = 'area', alpha=0.35, figsize =(20, 10))

ax.set_title('Immigration Trend of Top 5 Countries')
ax.set_ylabel('Number of Immigrants')
ax.set_xlabel('Years')
```

```python
df_least5 = df_can.tail(5)
df_least5 = df_least5[years].transpose()

df_least5.plot(kind='area', 
               alpha=0.45, 
               stacked=True, 
               figsize=(20, 10))

plt.title('Imigration trend of leat 5 countries')
plt.ylabel('Number of Imigrants')
plt.xlabel('Years')

plt.show()
```

```python
ax = df_least5.plot(kind='area', alpha=0.35, stacked=False,  figsize=(20, 10))

ax.set_title('Imigration trend of leat 5 countries')
ax.set_ylabel('Number of Imigrants')
ax.set_xlabel('Years')
```

## Histogram

```python
df_can['2013'].head()
```

```python
count, bin_edges = np.histogram(df_can['2013'])
print(count) #frekuensi
print(bin_edges) #bin ranges, default = 10 bins
```

```python
#membuat grafik distribusi
df_can['2013'].plot(kind='hist', figsize = (8, 5))

plt.title('Histogram of Immigration from 195 countries in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
# memperbaikin sumbu x agar cocok dengan ukuran bin dengan menggunakan xsticks
count, bin_edges= np.histogram(df_can['2013'])
df_can['2013'].plot(kind='hist', figsize=(8, 5), xticks=bin_edges)

plt.title('Histogram of imigration from 195 countries in 2013')
plt.ylabel('Number of countries')
plt.xlabel('Number of imigrants')

plt.show()
```

```python
df_can.loc[['Denmark', 'Norway', 'Sweden'], years]
```

```python
#buat ke dalam bentuk histogram
df_can.loc[['Denmark', 'Norway', 'Sweden'], years].plot.hist()
```

```python
df_t = df_can.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
df_t.head()
```

```python
df_t.plot(kind='hist', figsize=(10, 6))

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

```python
count, bin_edges = np.histogram(df_t, 15)

df_t.plot(kind = 'hist',
             figsize=(10, 6),
             bins=15,
             alpha=0.6,
             xticks=bin_edges,
             color=['coral', 'darkslateblue', 'mediumseagreen'])
plt.title('Histogram of imigration')
plt.ylabel('Number of years')
plt.xlabel('Number of imigrants')

plt.show()
```

```python
#melihat warna apasaja yang ada di matplotlib
import matplotlib
for name, hex in matplotlib.colors.cnames.items():
    print(name, hex)
```

```python
count, bin_edges = np.histogram(df_t, 15)

xmin = bin_edges[0] - 10
xmax = bin_edges[-1] + 10

df_t.plot(kind = 'hist',
          figsize=(10, 6),
          bins=15,
          xticks=bin_edges,
          stacked = True,
          xlim = (xmin, xmax),
          color=['coral', 'darkslateblue', 'mediumseagreen'])

plt.title('Histogram of imigration')
plt.ylabel('Number of years')
plt.xlabel('Number of imigrants')

plt.show()
```

```python
df_cof = df_can.loc[['Greece', 'Albania', 'Bulgaria'], years]
df_cof = df_cof.transpose()

count, bin_edges = np.histogram(df_cof, 15)
df_cof.plot(kind = 'hist',
              figsize=(10, 6),
              bins = 15,
              alpha = 0.35,
              xticks = bin_edges,
              color = ['coral', 'darkslateblue', 'mediumseagreen']
              )
plt.title('Histogram of Immigration from Greece, Albania, and Bulgaria from from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')
plt.show()
```
